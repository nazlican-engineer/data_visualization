```python
import seaborn as sns
tips=sns.load_dataset("tips")
df=tips.copy()
```

```python
df.info()
```

**TİPS VERİ SETİ**

✅total_bill:Bahşiş de dahil olmak üzere ödenen toplam ücret

✅tip:ödenen bahşiş

✅sex:cinsiyet

✅day:hangi günler restorana gelinmiş

✅time:günün hangi zamanında daha çok gelinmiş

✅size:gelenler toplam kaç kişi

```python
sns.barplot(x="day",y="total_bill",data=df);
```

```python
#BOXPLOT-->Bir değerin aykırı gözlem olduğunu bulmamız için  boxplot grafiği daha çok yardım eder
sns.boxplot(x=df["total_bill"]);#kutuda oluşan ortadaki çizgi medyanı gösterir
```

```python
df["total_bill"].median()
```

```python
df.groupby("day").total_bill.max()
```

```python
#hangi günler daha fazla kazanıyoruz
sns.boxplot(x="day",y="total_bill",data=df);
```

```python
#sabah mı akşam mı daha çok kazanıyoruz
sns.boxplot(x="time",y="total_bill",data=df)
```

```python
#kişi sayısı ile kazanç
sns.boxplot(x="size",y="total_bill",data=df)
```

```python
sns.boxplot(x="day",y="total_bill",hue="sex",data=df);
```

```python
#VİOLİN
sns.catplot(y="total_bill",kind="violin",data=df);
```

```python
sns.catplot(x="day",y="total_bill",kind="violin",data=df);
```

```python
sns.catplot(x="day",y="total_bill",kind="violin",hue="sex",data=df);
```
