```python
import seaborn as sns
fmri=sns.load_dataset("fmri")
df=fmri.copy()
```

```python
df.info()
```

**FMRİ VERİ SETİ**

✅subject:deney kimliğini gösterir

✅timepoint:ölçüm yapılan zamandaki zaman birimi

✅event:deneyin hangi durumlarda alındığını gösteriyor

✅region:beynin hangi bölgesinde ölçüm yapıldığını gösteriyor

✅signal:beynin tepkisini gösterir

```python
#LİNE PLOT
```

```python
sns.lineplot(x="timepoint",y="signal",data=df);
#mavi çizgi ortalamayı
#mavi çizginin etrafında oluşan mavi alanda standart sapmayı gösterir
```

```python

sns.lineplot(x="timepoint",y="signal",data=df,hue="event");
```

```python
sns.lineplot(x="timepoint",y="signal",data=df,hue="event",style="event");
```

```python
sns.lineplot(x="timepoint",
             y="signal",
             hue="event",
             style="event",
             markers=True,#belirli noktalara işaret koyması için
             dashes=False,#çizgilerin kesikli olmaması için
             data=df);
```

```python
sns.lineplot(x="timepoint",
             y="signal",
             hue="region",
             style="event",
             markers=True,#belirli noktalara işaret koyması için
             dashes=False,#çizgilerin kesikli olmaması için
             data=df);
```
