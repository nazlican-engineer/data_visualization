```python
#HEATMAP İÇERİSİNDE ZAMAN SERİSİ VERİLİERİ OLDUĞUNDA KULLANILIR.
import seaborn  as sns
flights=sns.load_dataset("flights")
df=flights.copy()
```

```python
df = df.pivot(index="month", columns="year", values="passengers")
```

```python
df
```

```python
sns.heatmap(df);#sütunlar sayısal olmalı
```

```python
sns.heatmap(df,annot=True,fmt="d");#fmt=d tamsayı türünde göster
```

```python
sns.heatmap(df,annot=True,fmt="d",linewidths=.5)#çizgilerin ölçüsünü ayarlamak için kullanılır
```

```python
sns.heatmap(df,annot=True,fmt="d",linewidths=.5,cbar=False);
```
