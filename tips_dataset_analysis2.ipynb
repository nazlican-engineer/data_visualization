```python
import seaborn as sns
tips=sns.load_dataset("tips")
df=tips.copy()
```

```python
df.info()
```

**TİPS VERİ SETİ**

✅total_bill:Bahşiş de dahil olmak üzere ödenen toplam ücret

✅tip:ödenen bahşiş

✅sex:cinsiyet

✅day:hangi günler restorana gelinmiş

✅time:günün hangi zamanında daha çok gelinmiş

✅size:gelenler toplam kaç kişi

```python
df.head()
```

```python
#KORELASYON GRAFİKLERİ
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df);#iki sayısal değişken arasındaki ilişkiyi göstermek için kullanılır.
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="time",data=df);
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="time",style="time",data=df);#style istediğim sütunun kategorilerini farklı şekillerde göstermek için kullanılır
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="day",style="time",data=df);
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="time",style="size",data=df);
```

```python
sns.scatterplot(x="total_bill",y="tip",size="size",hue="size",data=df);#size sayısal değişkenlerl kullanmak için kullanılır,renkli kullanmak için hue de kullanılabiliri
```

```python
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="total_bill",y="tip",data=df);#doğrusal bir çizgi kullanacağımız zaman bu grafiği kullanıyoruz lm-->linear model
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=df);
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",data=df);#col argümanı verilmiş olan kategorik değişkene göre ikiye böler
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",row="sex",data=df);
```
